```python
import csv
from os import walk
#반드시 반드시!!!! 파일 경로로 가서 이름 바꿔주기(behind the goal-->behind_the_goal)
f=open("train_game.csv",'w',newline='')
wr=csv.writer(f)
mypath="./videos/train/behind_the_goal"
for (dirpath, dirnames, filenames) in walk(mypath):
  for filename in filenames:
    dir="/content/drive/MyDrive/Colab Notebooks/game/videos/train/behind_the_goal/"+str(filename)+" 0"
    wr.writerow([dir])

mypath="./videos/train/beside_the_goal"
for (dirpath, dirnames, filenames) in walk(mypath):
  for filename in filenames:
    dir="/content/drive/MyDrive/Colab Notebooks/game/videos/train/beside_the_goal/"+str(filename)+" 1"
    wr.writerow([dir])

mypath="./videos/train/close-up_corner"
for (dirpath, dirnames, filenames) in walk(mypath):
  for filename in filenames:
    dir="/content/drive/MyDrive/Colab Notebooks/game/videos/train/close-up_corner/"+str(filename)+" 2"
    wr.writerow([dir])

mypath="./videos/train/close-up_player_or_referee"
for (dirpath, dirnames, filenames) in walk(mypath):
  for filename in filenames:
    dir="/content/drive/MyDrive/Colab Notebooks/game/videos/train/close-up_player_or_referee/"+str(filename)+" 3"
    wr.writerow([dir])

mypath="./videos/train/close-up_side_staff"
for (dirpath, dirnames, filenames) in walk(mypath):
  for filename in filenames:
    dir="/content/drive/MyDrive/Colab Notebooks/game/videos/train/close-up_side_staff/"+str(filename)+" 4"
    wr.writerow([dir])

mypath="./videos/train/crowd"
for (dirpath, dirnames, filenames) in walk(mypath):
  for filename in filenames:
    dir="/content/drive/MyDrive/Colab Notebooks/game/videos/train/crowd/"+str(filename)+" 5"
    wr.writerow([dir])

mypath="./videos/train/drone"
for (dirpath, dirnames, filenames) in walk(mypath):
  for filename in filenames:
    dir="/content/drive/MyDrive/Colab Notebooks/game/videos/train/drone/"+str(filename)+" 6"
    wr.writerow([dir])

mypath="./videos/train/main"
for (dirpath, dirnames, filenames) in walk(mypath):
  for filename in filenames:
    dir="/content/drive/MyDrive/Colab Notebooks/game/videos/train/main/"+str(filename)+" 7"
    wr.writerow([dir])

mypath="./videos/train/offside_line"
for (dirpath, dirnames, filenames) in walk(mypath):
  for filename in filenames:
    dir="/content/drive/MyDrive/Colab Notebooks/game/videos/train/offside_line/"+str(filename)+" 8"
    wr.writerow([dir])

mypath="./videos/train/player_back"
for (dirpath, dirnames, filenames) in walk(mypath):
  for filename in filenames:
    dir="/content/drive/MyDrive/Colab Notebooks/game/videos/train/player_back/"+str(filename)+" 9"
    wr.writerow([dir])

f.close()
```
